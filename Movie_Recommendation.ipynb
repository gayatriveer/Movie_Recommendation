# Creating a helper function
* '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14,
"name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
* converting this into list of genres
* ['Action','Adventure','name','Fantasy','Science Fiction']
* Here if you see closely content in genres colum is actually a string.
* ast.lieral_eval funtion is converting that string into list

```python
import ast

def convert(obj):
  list_1=[]
  for i in ast.literal_eval(obj):
   list_1.append(i['name'])
  return list_1

```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# We will use the same function to get list of keywords

movies['keywords'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
# For cast we will only select first 3 actors who will be the lead actors
# Slight changes in the convert function

def convert_3(obj):
  list_1=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter !=3:
      list_1.append(i['name'])
      counter+=1
    else:
      break
  return list_1
```

```python
movies['cast'].apply(convert_3)
```

```python
movies['cast']=movies['cast'].apply(convert_3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
#Now from the crew member we will just need the Director's name
# where the job is director
# Again, slight changes in the convert function

def fetch_director(obj):
  list_1=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      list_1.append(i['name'])
      break
  return list_1
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

# What is Stemming?

Stemming is the process of reducing words to their root or base form.

* Input: ['loving','loved', 'loves', 'lovely']
* Output: ['love','love','love', 'love']

```python
# Apply Stemming
import nltk
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
ps.stem('lovely')
```

```python
#Helper Function

def stemming(text):
  l=[]

  for i in text.split():
    l.append(ps.stem(i))

  string=" ".join(l)
  return string
```

```python
movies_new_df['tags'].apply(stemming)
```

```python
movies_new_df['tags']=movies_new_df['tags'].apply(stemming)
```

# Finding the nearest Vector (Similar movies) with cosine distance


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
# Function to suggest 5 similar movies

def recommend(movie):
  # Fetching index of the movie

  movie_index=movies_new_df[movies_new_df['title']== movie].index[0]
  distances= similarity[movie_index]

  # Used enumerate function so that our movie indices won't get lost while sorting
  # key in sorting-(4,0.345) we want to sort in 0.345(our distance from vector) not 4(4 is enumerate, index)

  movie_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x: x[1])[1:6]

  # print movie names

  for i in movie_list:
    print(movies_new_df.iloc[i[0]].title)

```

```python
recommend('Avatar')
```

```python
recommend("Pirates of the Caribbean: At World's End")
```
